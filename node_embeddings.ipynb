# SDNE

```
! git clone https://github.com/shenweichen/GraphEmbedding
%cd /content/GraphEmbedding/
!python setup.py install
from google.colab import drive
drive.mount('/content/drive')
import networkx as nx
from ge.classify import read_node_label,Classifier

from ge import SDNE

path = '/content/drive/My Drive/Altegrad_challenge/'

G_nx = nx.read_edgelist(path + 'collaboration_network.edgelist', delimiter=' ', nodetype=int)

model = SDNE(G_nx,hidden_size=[256,128]) #init model
model.train(batch_size=1,epochs=100,verbose=2)# train model
embeddings = model.get_embeddings()# get embedding vectors
```

# LINE

```
! git clone https://github.com/shenweichen/GraphEmbedding
%cd /content/GraphEmbedding/
!python setup.py install
from google.colab import drive
drive.mount('/content/drive')
import networkx as nx
from ge.classify import read_node_label,Classifier

from ge import LINE

path = '/content/drive/My Drive/Altegrad_challenge/'
G_nx = nx.read_edgelist(path + 'collaboration_network.edgelist', delimiter=' ', nodetype=int)


# change keep_dim to keepdim in line.py

model = LINE(G_nx,embedding_size=128,order='second') #init model,order can be ['first','second','all']
model.train(batch_size=10000,epochs=100,verbose=2)# train model
embeddings = model.get_embeddings()# get emb
```

```
embeddings = model.get_embeddings()# get emb

import pandas as pd
df = pd.DataFrame.from_dict(embeddings)
df = df.T
df.reset_index(inplace=True)
df
```

```
df.to_csv(path+'line_embed.csv')
```
